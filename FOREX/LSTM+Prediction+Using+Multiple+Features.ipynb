```python
#DEEP LEARNING MODEL
import numpy
import matplotlib as plt
import matplotlib.ticker as mticker
import matplotlib.dates as mdates
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import numpy as arange
import math
from sklearn.svm import SVR
from sklearn.metrics import explained_variance_score
from sklearn import linear_model
from keras.layers.core import Dense, Activation, Dropout
from keras.layers.recurrent import LSTM
from keras.models import Sequential
import lstm, time #helper libraries
from datetime import datetime
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error

#Retrieve data, extract features and send to classifer
#WITH PRICES FROM THE DATES ARTICLES WERE POSTED WE CAN TRY AND PREDICT FUTURE EVENTS
#RESEARCH SHOWS DEEP LEARNING METHODS TO BE MOST SUCCESSFUL
#LSTM MODEL
def main():

    #Retrieve data from CSV FILE
    stock = pd.read_csv("data/USDJPY.csv")

    #Organize attributes

    #REVERSE DATA SO IT STARTS FROM PAST
    stock = stock.reindex(index=stock.index[::-1])

    stock = stock[2:]

    frames = [stock]

    stock = pd.concat(frames)
    
    features(stock)
    
#GDP PRICES?
#INTEREST RATES?
#ADD FEATURES HERE
def features(stock):
    
    #GATHER DATES
    dates = stock["Date"]
    
    #SET DATE AS LABEL
    stock.set_index('Date', inplace=True)

    #GATHER ALL PRICES
    prices = stock["Price"]
    
    #GATHER OPEN PRICE
    openPrices = stock["Open"]
    
    #GATHER LOWEST PRICE
    lowPrices = stock["Low"]
    
    #GATHER LOWEST PRICE
    highPrices = stock["High"]
    
    #GATHER LOWEST PRICE
    changeP = stock["Change %"]
    
    #CONVERT ALL TO LISTS 
    prices = list(prices)
    
    dates = list(dates)
    days = []
    openPrices = list(openPrices)
    lowPrices = list(lowPrices)
    highPrices = list(highPrices)
    changeP = list(changeP)
    
    #ALL FEATURES ADDED TO LIST
    features = []
    datesConverted = []
    
    #Convert feature list and Labels to list
    for i in range(len(prices)):
        
        #Create Feature List
        features.append([float(prices[i]),float(openPrices[i]),float(lowPrices[i]),float(highPrices[i]),float(changeP[i])])
        
        #CONVERT DATES and Create List
        datesConverted.append(datetime.strptime(dates[i], "%b %d, %Y"))
        
        #days.append(i+1)
        
        
    lstmModel(prices, features)
    
#Create Dataset for LSTM Model
def create_dataset(dataset, look_back):
    dataX, dataY = [], []
    for i in range(len(dataset)-look_back-1):
        a = dataset[i:(i+look_back), :]
        dataX.append(a)
        dataY.append(dataset[i + look_back, :])
    return np.array(dataX), np.array(dataY)
    
def lstmModel(prices, features):
    
    lookback = 7
    
    #Normalize Prices
    scaler = MinMaxScaler(feature_range=(0, 1))
    prices = np.array(prices)
    prices = prices.reshape(len(prices), 1)
    prices = scaler.fit_transform(prices)
    
    #Normalize Features
    features = np.array(features)
    features = features.reshape(len(features), 5)
    features = scaler.fit_transform(features)
        
    # split into train and test sets
    train_size = int(len(features) * 0.80)
    test_size = len(features) - train_size
    train, test = features[0:train_size,:], features[train_size:len(features),:]
    
    #trainPrices
    trainPrices, testPrices = prices[0:train_size,:], prices[train_size:len(prices),:]
    
    # convert stock price data into time series dataset
    trainX, trainY = create_dataset(train, lookback)
    testX, testY = create_dataset(test, lookback)
    
    # reshape input of the LSTM to be format [samples, time steps, features]
    trainX = np.reshape(trainX, (trainX.shape[0], trainX.shape[1], 5))
    testX = np.reshape(testX, (testX.shape[0], testX.shape[1], 5))
    
    #Step 1 Build Model
    model = Sequential()
    
    # create and fit the LSTM network
    #3 LSTMs With 1 Dense
    model.add(LSTM(100, return_sequences=True, input_shape=(lookback,5), activation='tanh', recurrent_activation='hard_sigmoid'))
    model.add(LSTM(100, return_sequences=True))
    model.add(LSTM(100))
    model.add(Dense(5))
    #USE MSE LOSS FOR REGRESSION!
    model.compile(loss='mse', optimizer='adam', metrics=['accuracy'])
    model.fit(trainX, trainY, epochs=10000, batch_size=64, verbose=1)
    
    # make predictions
    trainPredict = model.predict(trainX)
    testPredict = model.predict(testX)
    
    #invert predictions and targets to unscaled
    trainPredict = scaler.inverse_transform(trainPredict)
    testPredict = scaler.inverse_transform(testPredict)
    trainY = scaler.inverse_transform(trainY)
    testY = scaler.inverse_transform(testY)
    
    # calculate root mean squared error
    trainScore = math.sqrt(mean_squared_error(trainY[:,0], trainPredict[:,0]))
    print('Train Score: %.2f RMSE' % (trainScore))
    testScore = math.sqrt(mean_squared_error(testY[:,0], testPredict[:,0]))
    print('Test Score: %.2f RMSE' % (testScore))
    
    # shift predictions of training data for plotting
    trainPredictPlot = np.empty_like(features)
    trainPredictPlot[:, :] = np.nan
    trainPredictPlot[lookback:len(trainPredict)+lookback, :] = trainPredict

    # shift predictions of test data for plotting
    testPredictPlot = np.empty_like(features)
    testPredictPlot[:, :] = np.nan
    testPredictPlot[len(trainPredict)+(lookback*2)+1:len(features)-1, :] = testPredict
    
    # plot baseline and predictions
    features = scaler.inverse_transform(features)
    #PLOT PRICE
    plt.plot(features[:,:1],color="blue")
    plt.plot(trainPredictPlot[:,:1],color="green")
    plt.plot(testPredictPlot[:,:1],color="red")
    plt.show()
    
    
main()
```
